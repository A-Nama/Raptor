```python
no=int(input("Enter a four digit number: "))

ones=no%10
thousands=no//1000
tens=(no//10)%10
hundreds=(no//100)%10
sum=ones+tens+hundreds+thousands
rev=(ones*1000)+(tens*100)+(hundreds*10)+thousands
diff=(thousands*tens)-(hundreds*ones)
print("The sum of it's numbers is ", sum)
print("The number's reverse is ", rev)
print("The difference between products of digits at odd and even positions is ", diff)
```

```python
def get():
  a=int(input("Enter first side: "))
  b=int(input("Enter second side: "))
  c=int(input("Enter third side: "))
  return a,b,c

def area(a,b,c):
  s=(a+b+c)/2
  A=(s*(s-a)*(s-b)*(s-c))**0.5
  return A
print("Enter values for first triangle: ")
a,b,c=get()
A1=area(a,b,c)
print("Enter values for second triangle: ")
d,e,f=get()
A2=area(d,e,f)
totalA=A1+A2
cA1=(A1/totalA)*100
cA2=(A2/totalA)*100
print("Total Area enclosed is ", totalA)
print("The contribution of first triangle is ", cA1)
print("The contribution of second triangle is ",cA2)
```

```python
def check(n):
  for i in range(1,101):
    sum=0
    while (n>0):
      r=n%10
      sum+=(r**2)
      n=n//10
    if (sum==1):
      return True

    elif(n<1):
      n=sum
    else:
      return False


def In_range(lower,upper):
  count=0
  for i in range(lower,upper):
    if check(i)==True:
      count+=1
      print(i)
  if count==0:
    print("There are no happy numbers in this range.")


def print_happy(x):
  count=0
  i=1
  while(count!=x):
    if check(i)==True:
      print(i)
      count+=1
    i+=1

    if count==x:
      break

n=int(input("Enter the number you want to check: "))
if (check(n)==True):
  print("It's a happy number.")
else:
  print("It's a sad number.")

lower=int(input("Enter the lower limit of range: "))
upper=int(input("Enter the upper limit of range: "))
In_range(lower,upper)

x=int(input("Enter the number of happy numbers you want printed: "))
print_happy(x)
```

```python
def print_all_substrings(s):
    n = len(s)
    for i in range(n):
        for j in range(i+1,n+1):
            print(s[i:j])

def print_substrings_of_length_k(s, k):
    n = len(s)
    if k<=0 or k>n:
        print("Invalid length.")
        return

    for i in range(n-k+1):
        substring = s[i:i +k]
        print(substring)

def print_substrings_length_k_distinct_chars(s, k):
    n = len(s)
    if k<=0 or k>n:
        print("Invalid length.")
        return

    for i in range(n-k+1):
        substring = s[i:i +k]
        if len(set(substring)) == k:
            print(substring)

def print_max_length_substrings_distinct_chars(s, n):
    max_length = 0
    max_substrings = []

    for i in range(len(s)):
        chars_set = set()
        current_length = 0
        current_substring = ""

        for j in range(i, len(s)):
            if s[j] not in chars_set:
                chars_set.add(s[j])
                current_length += 1
                current_substring += s[j]
            else:
                break

        if current_length > max_length:
            max_length = current_length
            max_substrings = [current_substring]
        elif current_length == max_length:
            max_substrings.append(current_substring)

    print("Maximum length substrings with {} distinct characters:".format(n))
    for substring in max_substrings:
        print(substring)

def is_palindrome(s):
    return s == s[::-1]

def print_palindrome_substrings(s):
    n = len(s)
    for i in range(n):
        for j in range(i + 1, n + 1):
            substring = s[i:j]
            if is_palindrome(substring):
                print(substring)


s = str(input("Enter a string: "))

print("All possible substrings:")
print_all_substrings(s)

k = int(input("Enter the length of substrings (K): "))
print(f"All possible substrings of length {k}:")
print_substrings_of_length_k(s, k)

n = int(input("Enter the length of substrings (K) with N distinct characters: "))
print(f"All substrings of length {n} with {n} distinct characters:")
print_substrings_length_k_distinct_chars(s, n)

d = int(input("Enter the number of distinct characters (N): "))
print_max_length_substrings_distinct_chars(s, d)

print("All palindrome substrings:")
print_palindrome_substrings(s)
```
